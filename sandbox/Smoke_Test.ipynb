## Test add_sphere and add_spheres

```python
v = viewer.MolSysView()

v.load_pdb_string(pdb_text)
v.shapes.add_sphere(
    center=(12.5, 13.0, 8.3),
    radius=2.0,
    color=0x00ff00,
    alpha=0.4,
    tag="test_sphere_1",
)
v.show()
```
